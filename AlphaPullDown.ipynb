```python
import os
import json
import pandas as pd

# æ§‹é€ äºˆæ¸¬çµæžœã®ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªï¼ˆé©å®œå¤‰æ›´ã—ã¦ãã ã•ã„ï¼‰
results_dir = "/content/prediction_results2"

# CSV å‡ºåŠ›å…ˆãƒ•ã‚¡ã‚¤ãƒ«å
output_csv = os.path.join(results_dir, "predictions_with_good_interpae.csv")

rows = []

# results_dir å†…ã®å„ã‚µãƒ–ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªï¼ˆå„è¤‡åˆä½“ã®çµæžœãƒ•ã‚©ãƒ«ãƒ€ï¼‰ã‚’èµ°æŸ»
for subdir in os.listdir(results_dir):
    sub_path = os.path.join(results_dir, subdir)
    if os.path.isdir(sub_path):
        ranking_debug_path = os.path.join(sub_path, "ranking_debug.json")
        if os.path.exists(ranking_debug_path):
            with open(ranking_debug_path, "r") as f:
                data = json.load(f)

            # JSON ã«å«ã¾ã‚Œã‚‹ 'order' ã‚­ãƒ¼ã‹ã‚‰ã€æœ€ã‚‚ä¿¡é ¼æ€§ã®é«˜ã„ãƒ¢ãƒ‡ãƒ«ã‚’é¸æŠž
            order = data.get("order", [])
            if order:
                best_model = order[0]
                iptm = data.get("iptm", {}).get(best_model, None)
                iptm_plus_ptm = data.get("iptm+ptm", {}).get(best_model, None)

                # ptm ã¯ (iptm+ptm) - iptm ã§è¨ˆç®—
                ptm = iptm_plus_ptm - iptm if (iptm is not None and iptm_plus_ptm is not None) else None

                # mpDockQ, PI_score ãŒå­˜åœ¨ã™ã‚Œã°ãã‚Œãžã‚Œå–å¾—ï¼ˆå­˜åœ¨ã—ãªã„å ´åˆã¯ Noneï¼‰
                mpDockQ = data.get("mpDockQ", {}).get(best_model, None)
                PI_score = data.get("PI_score", {}).get(best_model, None)
            else:
                iptm = None
                iptm_plus_ptm = None
                ptm = None
                mpDockQ = None
                PI_score = None

            rows.append({
                "complex": subdir,
                "iptm": iptm,
                "ptm": ptm,
                "iptm+ptm": iptm_plus_ptm,
                "mpDockQ": mpDockQ,
                "PI_score": PI_score
            })

# pandas DataFrame ã«å¤‰æ›ã—ã¦ CSV ã¨ã—ã¦ä¿å­˜
df = pd.DataFrame(rows)
df.to_csv(output_csv, index=False)

print("CSV ãƒ•ã‚¡ã‚¤ãƒ«ãŒä½œæˆã•ã‚Œã¾ã—ãŸï¼š", output_csv)
```
